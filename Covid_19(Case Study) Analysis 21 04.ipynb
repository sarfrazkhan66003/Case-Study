#### Find maximum infection rate for chine

##### Find maximum infection rate for all of the countries

```python
countries = list(coronna_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(coronna_dataset_aggregated.loc[c].diff().max())

max_infection_rates
```

```python
coronna_dataset_aggregated['max_infection_rates'] = max_infection_rates
```

```python
coronna_dataset_aggregated.head()
```

```python
#create a new dataframe with only needed column
coronna_data = pd.DataFrame(coronna_dataset_aggregated['max_infection_rates'])
```

```python
coronna_data.head()
```

#### Happiness DataSet

```python
happiness_report_csv = pd.read_csv('worldwide_happiness_report.csv')
```

```python
happiness_report_csv.head()
```

#### Changing the indices of the dataframe

```python
happiness_report_csv.set_index('Country or region', inplace = True)
happiness_report_csv.head()
```

#### Now let's join two dataset we have prepared

```python
#inner join
data = coronna_data.join(happiness_report_csv,how = 'inner')
data.head()
```
